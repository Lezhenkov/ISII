## Часть 1: генерация одномерного массива с помощью numpy и встроенных средств python

Для данной задачи при выводе массива ограничимся 10 элементами: 5 первых и 5 последних. При обычном выводе возникла проблема с google colab. При попытке напечатать массив в 100.000 элементов с помощью встроенных средств: система сообщала о перегрузке памяти консоли. В связи с этим и для удобства восприятия и были наложены ограничения

```python
def gener_d1_massive_np():
  # Пользователь вводит исходные данные:
  x, a, b = int(input("введите число элементов массива ")), int(
      input('введите левую границу массива ')), int(input('введите правую границу массива '))
  tic = time.perf_counter_ns()  # запуск таймера
  # Добавляем элементы в массив. Данное выражение требуется для учета границ, введенных пользователем
  d1 = (b-a) * np.random.sample(x) + a 
  tac = time.perf_counter_ns() # остановка таймера
  # для удобства чтения и защиты от ошибок вводим условие на вывод массива, чтобы не засорять консоль ненужными данными
  if len(d1) > 10:
    print("Random number list is : ", d1[0:5], '...', d1[-6:-1], sep = '\n')
  else:
    print ("Random number list is : ", d1, sep = '\n')  # выводим массив пользователю
  # находим и выводим время, затраченное на генерацию массива в наносекундах 
  print(f"Создание массива заняло {tac-tic} нсек") 
  pass


def gener_d1_massive_np_poor():
  # Пользователь вводит исходные данные:
  x = int(input("введите число элементов массива "))
  tic = time.perf_counter_ns()  # записываем первое значение таймера
  # Добавляем элементы в массив. Данное выражение требуется для учета границ, введенных пользователем
  d1 = np.random.sample(x)
  tac = time.perf_counter_ns() # записываем второе
  # цикл для ограничения вывода
  if len(d1) > 10:
    print("Random number list is : ", d1[0:5], '...', d1[-6:-1], sep = '\n')
  else:
    print ("Random number list is : ", d1, sep = '\n')  # выводим массив пользователю
  # находим и выводим время, затраченное на генерацию массива в наносекундах 
  print(f"Создание массива заняло {tac-tic} нсек") 
  pass


def gener_d1_massive_py():
  # пользователь вводит исходные данные
  x, a, b = int(input("введите число элементов массива ")), int(
      input('введите левую границу массива ')), int(input('введите правую границу массива '))
  tic = time.perf_counter_ns() # старт таймера
  d1 = [] # создаем пустой массив, куда будем добавлять элементы 
  # запускаем цикл по добавлению элементов 
  for i in range(x):   # диапазоном для цикла является введеный пользователем x - длина массива
    d1.append(random.uniform(a,b)) # внутри цикла добавляем элементы по одному, указывая границы генерации случайного числа
  tac = time.perf_counter_ns()
  if len(d1) > 10:
    print("Random number list is : ", d1[0:5], '...', d1[-6:-1], sep = '\n')
  else:
    print ("Random number list is : ", d1, sep = '\n')  # выводим массив пользователю
  print(f"Создание массива заняло {tac-tic} нсек")
  pass
```

```python
print(gener_d1_massive_np())
```

```python
print(gener_d1_massive_np_poor())
```

```python
print(gener_d1_massive_py())
```

Следует обратить внимание, что при генерации с помощью numpy было использовано 2 функции. Разница заключается в том, что функция gener_d1_massive_np() имеет возможность ручного задавания границ, а gener_d1_massive_np_poor() данной возможности не имеет, а выводит числа в диапозоне [0,1). В результате вторая функция работает в 2 раза быстрее, т.к. ей не требуется проводить дополнительное вычисление выражения внутри добавления элемента в массив. 

Если сравнивать одинаковые по реализации функции с numpy и встроенными средствами python, то время отличается в разной степени. Встроенные средства python выигрывают по времени на коротких выборках, но проигрывают на длинных.

## Часть 2: генерация двумерного массива с помощью numpy и встроенных средств python

будем сравнивать 2 функции с одинаковыми условиями: диапазон генерации случайных чисел: [0, 1), т.е. дефолтные условия функции

Данные функции можно использовать для генерации как двумерых массивов, так и для n - мерных.

```python
def gener_d2_massive_np():
  n = [0, 0]
  n[0], n[1] = int(input(
      'введите количество элементов массива ')), int(input(
          'введите размерность массива '))
  tic = time.perf_counter_ns() # старт таймера
  # n подает в функцию 2 числа: первое - длина массива, второе - n-мерность массива
  d2 = np.random.sample(n)
  tac = time.perf_counter_ns()
  if len(d2) > 10:
    print("Random number list is : ", d2[0:5], '...', d2[-6:-1], sep = '\n')
  else:
    print ("Random number list is : ", d2, sep = '\n')  # выводим массив пользователю
  print(f"Создание массива заняло {tac-tic} нсек")
  return d2
```

```python
def gener_d2_massive_py():
  n = [0, 0]
  n[0], n[1] = int(input('введите количество элементов массива ')), int(
      input('введите размерность массива '))
  tic = time.perf_counter_ns() # старт таймера
  d2 = []
  for i in range(n[0]):
    d2.append([])
    for j in range(n[1]):
      d2[i].append(random.uniform(0,1)) 
      """можно указать или задать первоначальные  числа для генерации случайных чисел,
         но по условию данной части этого делать не будем"""
  tac = time.perf_counter_ns() # 
  if len(d2) > 10:
    print("Random number list is : ", d2[0:5], '...', d2[-6:-1], sep = '\n')
  else:
    print ("Random number list is : ", d2, sep = '\n')  # выводим массив пользователю
  print(f"Создание массива заняло {tac-tic} нсек")
  return d2
```
