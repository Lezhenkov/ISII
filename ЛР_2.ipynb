# 4. Вывод наименований тех средств, которые имеют просроченные сертификаты (на момент выполнения задания).

Чтобы успешно работать с датами, с помощью пандас сменим формат столбца на datetime64. В столбце присутствуют сроки без указанной даты, будем считать их истекшими.

```python
df_4 = df
# конвертируем строку в дату
df_4['срок действия'] = pd.to_datetime(df_4['срок действия'], format='%d.%m.%Y', errors='coerce')
# вводим сравнение даты с текущей датой
df_4 = df_4.loc[(df_4['срок действия'] >= pd.Timestamp.now())]
df_4
```

проверяем даты

```python
pd.unique(df_4['срок действия'])
```

# 5. Показать наименования тех средств, которые имеют просроченные сертификаты, но при этом обязаны оказывать поддержку (столбец «Информация об окончании срока технической поддержки, полученная от заявителя»)

в столбце конец срока поддержки, помимо дат имеется незаполненные значения "NaN" и "бессрочно". Если значение не указано, то будем считать, что срок тех.поддержки истек.

```python
df_5 = df
df_5.dropna(subset=['конец срока поддержки'], inplace = True) # сразу убираем значения NaN
df_5
```

```python
# заменим строку бессрочно на дату (условно зададим ее очень высокой, чтобы потом не изменить исходные данные)
df_5 = df_5.replace({'конец срока поддержки': {'бессрочно':'22.12.2222'}})
# Преобразуем данные в формат дат
df_5['конец срока поддержки'] = pd.to_datetime(df_5['конец срока поддержки'], format='%d.%m.%Y', errors='raise') 
# вводим условие задания
df_5 = df_5.loc[(df_5['срок действия'] <= pd.Timestamp.now()) & (df_5['конец срока поддержки'] >= pd.Timestamp.now())]
# преобразуем обратно дату в строку
df_5 = df_5.replace({'конец срока поддержки': {'22.12.2222': 'бессрочно'}})
```
